# Linkedin Web Scraper
### Goal:
- given: acquired skills, not acquired skills, job type
- output: jobs in order of best to worst match

#### Utility Functions

```python
from bs4 import BeautifulSoup
import urllib
import pandas as pd
```

```python
JOB_SEARCH_URL = "https://www.linkedin.com/jobs/search/?f_E=1&geoId=103644278&keywords=data&location=United%20States"
```

```python
# get source code of the page
def get_url(url):
    return urllib.request.urlopen(url).read()

# makes the source tree format like 
def beautify(url):
    source = get_url(url)
    return BeautifulSoup(source, "html.parser")

# get job id from job posting raw html
def get_job_id(job_posting_html):
    return str(job_posting_html).split(' ')[7].split(':')[-1]
```

#### Scraping Job Listings
- Each job posting in a list has multiple classes: `base-card base-card--link base-search-card base-search-card--link job-search-card job-search-card--active`, each one separated by a space.
- Each separate job posting has an ID, which needs to be affixed to the base URL as so: `JOB_SEARCH_URL + "&currentJobId=000"`
    - Job ID can be found inside the first `div` with `class=data-entity-urn`
-------
- from each job posting, we need the title, company, details, and date posted.
    - title can be found inside `<span class="screen-reader-text">`
    - company can be found inside `<a class="hidden-nested-link"...>`

```python
jobs = beautify(JOB_SEARCH_URL)
list_of_divs = jobs.find_all('div', {'class' : 'base-card'})
```
