# Linkedin Web Scraper
### Goal:
- given: acquired skills, not axquired skills, job type
- output: jobs in order of best to worst match

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import time
```

#### Logging In

```python
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning) 
```

```python
driver = webdriver.Chrome()

email, pwd = "vivrd@berkeley.edu", "!GalaxyS4"
driver.get('https://www.linkedin.com/login')
print("Login page found")
time.sleep(3)

driver.find_element_by_id('username').send_keys(email)
print("username sent")
driver.find_element_by_id('password').send_keys(pwd)
print("password sent")
driver.find_element_by_id('password').send_keys(Keys.RETURN)
```

#### Searching Jobs & Details

```python
driver.get("https://www.linkedin.com/jobs/")
time.sleep(3)

query = "data"
search_bar = driver.find_element_by_id("jobs-search-box-keyword-id-ember32")
search_bar.send_keys(query)
search_bar.send_keys(Keys.RETURN)
```
